```python
from telethon import TelegramClient
from telethon.tl import functions
import asyncio

# Replace with your API ID and API hash
api_id = 21722549  # Replace with your API ID
api_hash = ''  # Replace with your API hash

client = TelegramClient('anon', api_id, api_hash)

async def search_public_groups(keyword):
    result = await client(functions.contacts.SearchRequest(
        q=keyword,
        limit=100   # כמה תוצאות מקסימום
    ))

    for chat in result.chats:
        print(chat.id, getattr(chat, "title", None), getattr(chat, "username", None))

async def main():
    async with client:
        await search_public_groups("מיכה")

await main()
```
